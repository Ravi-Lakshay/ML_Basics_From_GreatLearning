```python
from matplotlib import pyplot as plt # here plt is alias for pyplot submodule
```

```python
import numpy as np
```

```python
x=np.arange(1,11)
x
```

```python
y = 2*x
y2=3*x
y.shape
```

```python
plt.plot(x,y,color='green',linewidth=15,linestyle=":") 
plt.plot(x,y2,color='blue',linewidth=15,linestyle=":")
plt.title("Line Plot") 
plt.xlabel('x-axis')
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# bar-plot, for this we first create a dictionary, # this plt.bar takes only list values!
```

```python
students = {'sam':30, 'Bob':50, 'Julia':70}
```

```python
names = list(students.keys())
marks = list(students.values())
```

```python
names # these are catgorical values 
```

```python
names, marks
```

```python
plt.bar(names,marks, color='orange')
plt.barh(names,marks, color='blue')
plt.title('marks of students')
plt.ylabel('Marks')
plt.xlabel('Names of students')
plt.grid(True)
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
iris = pd.read_csv('wine.csv', names = ["cultivator", "Alcohol", "Malic_acid", "Ash", "Alcalinity_of_ash", "Magnesium", "Total_phenols", "Flavanoids", "Nonflavanoid_phenols", "Proanthocyanins", "Color_intensity", "Hue", "OD280", "Proline"])
iris.head()
```

```python
plt.hist(iris['Flavanoids'], bins = 20)
plt.show()
```

```python
iris.boxplot(column='Flavanoids', by='cultivator')
```

```python
import seaborn as sns # for an even beautiful boxplot!
```

```python
fruits = ['apple', "mange","orange", 'banana']
cost = [76,45,90,85]
```

```python
plt.pie(cost,labels=fruits)  # in this first map the numerical values, and then map the labels attribute!
```

```python
plt.figure(figsize=(10,4))
plt.pie(cost,labels=fruits, autopct='%0.4f%%',shadow=True)
plt.show()
```
