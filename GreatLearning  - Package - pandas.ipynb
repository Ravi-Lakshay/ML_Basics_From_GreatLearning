Similarly a series object from a dictionary!

```python
d1 = {'a':10,'2':"sd", 'c':30}
```

```python
s1 = pd.Series(d1) # here the keys become the index values!
s1
```

```python
s1 = pd.Series(d1)
Index = ['a', 'b', 'c']
s1 = pd.Series(s1, Index ) # strangely we get NaN value!, may be if numbers are used then it will be normal!
s1
```

```python
d2 = {'a':10,'b':20, 'c':30}
s1 = pd.Series(d2)
Index = ['a', 'b', 'c']
s1 = pd.Series(s1, Index ) # now we get int64
s1
```

```python
d2 = {'a':10,'22':20, 'c':30}
s1 = pd.Series(d2)
Index = ['a', 'b', 'c']
s1 = pd.Series(s1, Index ) # now we get float64 and Nan
s1
```

```python
d2 = {'a':10,'b':"sdf", 'c':30}
s1 = pd.Series(d2)
Index = ['a', 'b', 'c']
s1 = pd.Series(s1, Index ) # now normal, and object and not int64
s1
```
