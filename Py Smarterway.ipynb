```python
#                               Chapter 18 Lists:  Deleting and removing elements
```

```python
tasks = ["email Frank", "call Sarah", "meet with Zach"]
del tasks[1]
tasks
```

```python
# OR
tasks = ["email Frank", "call Sarah", "meet with Zach"]
tasks.remove("call Sarah") # or tasks.remove(tasks[1])
tasks
```

```python
# OR
tasks = ["email Frank", "call Sarah", "meet with Zach"]
tasks.remove(tasks[0:2]) # or tasks.remove(tasks[1])
tasks
```

```python
# OR
tasks = ["email Frank", "call Sarah", "meet with Zach"]
del tasks[0:2] 
tasks
```
